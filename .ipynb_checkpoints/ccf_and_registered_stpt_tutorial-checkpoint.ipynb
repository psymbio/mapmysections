## Allen Mouse Common Coordinate Framework (2020 version)

Allen Mouse Brain Common Coordinate Framework (CCFv3, Wang et al, 2020) is a 3D reference space is an average brain at 10um voxel resolution created from serial two-photon tomography images of 1,675 young adult C57Bl6/J mice. Using multimodal reference data, the entire brain parcellated directly in 3D, labeling every voxel with a brain structure spanning 43 isocortical areas and their layers, 314 subcortical gray matter structures, 81 fiber tracts, and 8 ventricular structures. The 2020 version adds new annotations for layers of the Ammonâ€™s horn (CA), main olfactory bulb (MOB) and minor modification of surrounding fiber tracts.

CCFv3 is used in informatics pipelines and online applications to analyze, visualize and integrate multimodal and multiscale data sets in 3D, and is openly accessible for research use.

Information on how to download and use CCFv3 template, annotation and ontology can be found at the abc_atlas_access repo: https://alleninstitute.github.io/abc_atlas_access/descriptions/Allen-CCF-2020.html

For this tutorial, a subset of needed data files can be found in the S3 location below:
* average_template_25.nii.gz
* annotation_25.nii.gz
* annotation_boundary_25.nii.gz
* parcellation_to_parcellation_term_membership_acronym.csv
* parcellation_term.csv

https://map-my-sections-802451596237-us-west-2.s3.us-west-2.amazonaws.com/index.html#ccf-reference/

In this tutorial, we assume that you have downloaded the volumes to local directory downloaded_data/reference

```python
reference_path = os.path.join( 'downloaded_data','reference')
```

The "average_template_25" volume is the anatomical template of the CCF constructed as the shape and intensity average of 1675 specime brains. The "annotation_25" volume is the parcellation of the brain with respect to a heirachical partonomy of anatomical structures. The "annotation_boundary_25" volume is a mask that identifies all the voxel with on the boundary of a parcellation to support data visualization. The volumes are stored in compressed NIFTI (.nii.gz) format. In this notebook uses the SimpleITK library to open the volume.

```python
file = os.path.join( reference_path, 'average_template_25.nii.gz')
average_template_image = sitk.ReadImage(file)
average_template_array = sitk.GetArrayViewFromImage(average_template_image)

file = os.path.join( reference_path, 'annotation_25.nii.gz')
annotation_image = sitk.ReadImage(file)
annotation_array = sitk.GetArrayViewFromImage(annotation_image)
annotation_index = np.where( annotation_array > 0 )
annotation = annotation_array[annotation_index]
parcellation_index = np.unique( annotation_array[annotation_index] )

file = os.path.join( reference_path, 'annotation_boundary_25.nii.gz')
annotation_boundary_image = sitk.ReadImage(file)
annotation_boundary_array = sitk.GetArrayViewFromImage(annotation_boundary_image)
```

We define a helper function to visualize the sanme coronal section of the average template, annotation and boundary volumes

```python
def plot_section(slice, cmap = plt.cm.Greys_r, fig_width = 8, fig_height = 8):
    fig, ax = plt.subplots()
    fig.set_size_inches(fig_width, fig_height)
    if cmap is not None:
        plt.imshow(slice, cmap=cmap)
    else :
        plt.imshow(slice)
    plt.axis("off")
    return fig, ax
```

```python
zindex = 273
zslice = np.transpose(average_template_array[:, :, zindex])
fig, ax = plot_section(zslice)
res = ax.set_title('average_template_25')
plt.show()
```

```python
zslice = np.transpose(annotation_array[:, :, zindex])
fig, ax = plot_section(zslice)
res = ax.set_title('annotation_25')
plt.show()
```

```python
zslice = np.transpose(annotation_boundary_array[:, :, zindex])
fig, ax = plot_section(zslice, cmap=plt.cm.Greys)
res = ax.set_title('annotation_boundary_25')
plt.show()
```

## Summarizing intensity by structure

To summarize intensity by structure we need to open the following files:
* parcellation_to_parcellation_term_membership_acronym.csv : maps values in the annotation to anatomical structures at the division, structure and sub-structure levels
* parcellation_term.csv  : provides information about structure name, identifiers and colors


```python
file = os.path.join( reference_path, 'parcellation_term.csv')
term = pd.read_csv(file)
cols = ['acronym','name','identifier','color_hex_triplet']
term = term[cols].drop_duplicates()
term.set_index('acronym',inplace=True)
term.head(5)
```

```python
file = os.path.join( reference_path, 'parcellation_to_parcellation_term_membership_acronym.csv')
pivot = pd.read_csv( file ).set_index('parcellation_index')
pivot
```

We define a helper function that summarizes intensity statistics (count, min, max, sum, sum_squared, mean) for each unique label in the annotation volume. These values are then aggregated to define structural summaries at the division, structure and sub-structure levels.

```python
def summarize_intensity( experiment_dir, pivot ) :
    
    summary = {}
    summary['parcellation'] = {}

    for cc in ['red','green'] :

        print(cc)

        summary['parcellation'][cc] = pivot.copy()

        file = os.path.join( experiment_dir, 'resampled_%s_25.nii.gz' % cc )
        intImage = sitk.ReadImage( file )
        intArray = sitk.GetArrayFromImage( intImage )
        values = intArray[annotation_index]
        sqValues = np.square( values )

        for pidx in parcellation_index :

            labIndex = np.where( annotation == pidx )
            fValues = values[labIndex]
            sValues = sqValues[labIndex]
            if (fValues.size) :
                summary['parcellation'][cc].loc[pidx,'count'] = fValues.size
                summary['parcellation'][cc].loc[pidx,'min'] = fValues.min()
                summary['parcellation'][cc].loc[pidx,'max'] = fValues.max()
                summary['parcellation'][cc].loc[pidx,'sum'] = fValues.sum()
                summary['parcellation'][cc].loc[pidx,'sum_squared'] = sValues.sum()
                summary['parcellation'][cc].loc[pidx,'mean'] = fValues.mean()


        pred = pd.notna( summary['parcellation'][cc]['count'] )
        summary['parcellation'][cc] = summary['parcellation'][cc][pred].copy()
    
    for lvl in ['division','structure','substructure'] :

        summary[lvl] = {}

        for cc in ['red','green'] :

            print(lvl,cc)

            if lvl == 'substructure' :
                cols = ['organ','category','division','structure']
            elif lvl == 'structure' :
                cols = ['organ','category','division']
            elif lvl == 'division' :
                cols = ['organ','category']
            else:
                cols = []

            grouped = summary['parcellation'][cc].groupby([lvl])[cols].first()
            grouped_summed = summary['parcellation'][cc].groupby([lvl])[['count','sum','sum_squared']].sum()
            grouped_max = summary['parcellation'][cc].groupby([lvl])[['max']].max()
            grouped_min = summary['parcellation'][cc].groupby([lvl])[['min']].min()

            joined = grouped.join( grouped_summed )
            joined = joined.join( grouped_max )
            joined = joined.join( grouped_min )

            pred = joined['count'] > 0
            joined.loc[pred,'mean'] = joined.loc[pred,'sum'] / joined.loc[pred,'count']

            summary[lvl][cc] = joined
            
    return summary
```

```python
summary = summarize_intensity( example_path, pivot)
```

Here we show the top 20 red intensity structures for image-series 1329442699

```python
top20 = summary['structure']['red'].sort_values('mean',ascending=False).head(20)
top20 = top20.join(term[['name']])
top20
```
