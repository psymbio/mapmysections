```python
from cloudpathlib import CloudPath
from pathlib import Path

# Replace this with your actual S3 bucket link
s3_path = "s3://allen-genetic-tools/tissuecyte/1267695363/ome_zarr_conversion/1267695363.zarr/"

# Create a CloudPath object pointing to the S3 location
cloud_path = CloudPath(s3_path)

# Local directory to store the files
local_dir = Path("stpt_files")
local_dir.mkdir(exist_ok=True)

# Download all files from the S3 path
for file in cloud_path.iterdir():
    if file.is_file():
        local_file_path = local_dir / file.name
        file.download_to(local_file_path)
        print(f"Downloaded {file.name} to {local_file_path}?")
```
